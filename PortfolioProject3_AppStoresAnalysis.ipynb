# Creating outlier mask

```python
Q1 = apple['rating_count_tot'].quantile(0.25)
Q3 = apple['rating_count_tot'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

upper_mask = apple['rating_count_tot'] >= upper
lower_mask = apple['rating_count_tot'] <= lower

outlier_mask = ~(upper_mask | lower_mask)
apple_mask = apple[outlier_mask]
apple_mask.sort_values(by='rating_count_tot',ascending = False).head()
```

```python
Q1 = google['Installs'].quantile(0.25)
Q3 = google['Installs'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

upper_mask = google['Installs'] >= upper
lower_mask = google['Installs'] <= lower

outlier_mask = ~(upper_mask | lower_mask)
google_mask = google[outlier_mask]
google_mask.sort_values(by='Installs',ascending = False).head()
```

# Looking at top 5 apps from top 5 average most used genres

```python
apple_most_popular_genres = apple_mask.groupby('prime_genre')['rating_count_tot'].mean().sort_values(ascending=False).head()
google_most_popular_genres = google_mask.groupby('Category')['Installs'].mean().round().sort_values(ascending=False).head()
```

```python
for genre in apple_most_popular_genres.index:
    print(f"\nTop 5 apps in {genre} genre:")
    top_apps = apple_mask[apple_mask['prime_genre'] == genre]\
                 .sort_values(by='rating_count_tot', ascending=False)\
                 .head()
    display(top_apps)
```

```python
for genre in google_most_popular_genres.index:
    print(f"\nTop 5 apps in {genre} genre:")
    top_apps = google_mask[google_mask['Category'] == genre]\
                 .sort_values(by='Installs', ascending=False)\
                 .head()
    display(top_apps)
```
